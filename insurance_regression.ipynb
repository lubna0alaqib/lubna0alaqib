```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset
data = pd.read_csv('insurance.csv')
```

```python
# Plot a scatter plot with age on the X-axis and charges on the Y-axis
plt.scatter(data['age'], data['charges'])
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs. Insurance Charges')
plt.show()
```

```python
# Create a linear regression model
model = LinearRegression()

# Prepare the data
X = data['age'].values.reshape(-1, 1)
y = data['charges'].values

# Fit the model to the data
model.fit(X, y)
```

```python
# Make predictions
predictions = model.predict(X)

# Plot the scatter plot
plt.scatter(data['age'], data['charges'], label='Actual data')

# Plot the best-fit line
plt.plot(data['age'], predictions, color='red', label='Best-fit line')

plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs. Insurance Charges with Best-fit Line')
plt.legend()
plt.show()
```
